***Circular Reference and Garbage Collection***

```python
import ctypes
import gc
```

```python
def ref_count(address):
    return ctypes.c_long.from_address(address).value
```

```python
# Looking for the object on memory by GC
def object_by_id(object_id): # object's address location
    for obj in gc.get_objects():
        if id(obj) == object_id:
            return "object exits"
    return "Not Found"
```

```python
class A: # A point B
    def __init__(self):
        self.b = B(self)
        print("A: {0}, b: {1}".format(id(self), id(self.b)))
        

class B: # B point A
    def __init__(self, a):
        self.a = a
        print("B: {0}, a: {1}".format(id(self), id(self.a)))
```

```python
gc.disable() # to disable few functionality
```

```python
my_var1 = A()
```

```python
id(my_var1)
ref_count(id(my_var1))
a_id = id(my_var1)
b_id = id(my_var1.b)
print(a_id)
print(b_id)
```

```python
print(object_by_id(a_id))
print(object_by_id(b_id))
```

```python
my_var1 = None
```

```python
print(my_var1)
```

```python
print(object_by_id(a_id))
print(object_by_id(b_id))
```

```python
gc.collect()
```

```python
print(object_by_id(a_id))
print(object_by_id(b_id))
```
